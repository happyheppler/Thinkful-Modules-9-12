### 1. How many movies are released for each rating?

```python
# Construct the SQL statement
sql = '''
SELECT COUNT(*) AS count,
    rating
FROM film
GROUP BY rating
'''

results = engine.execute(sql)
engine.dispose()
rows = results.fetchall()

# Extract the counts and print results
counts = [row['count'] for row in rows]
print('There are {} movies rated R.'.format(counts[0]))
print('There are {} movies rated NC-17.'.format(counts[1]))
print('There are {} movies rated PG.'.format(counts[2]))
print('There are {} movies rated PG-13.'.format(counts[3]))
print('There are {} movies rated G.'.format(counts[4]))
```

### 2. What is the average rental duration for each rating?

```python
# Construct the SQL statement
sql = '''
SELECT AVG(rental_rate) AS avg_rental_rate,
    rating
FROM film
GROUP BY rating
'''

results = engine.execute(sql)
engine.dispose()
rows = results.fetchall()

# Extract the counts and print results
average = [row['avg_rental_rate'] for row in rows]
print('The average rental duration for movies rated R is {}.'.format(average[0]))
print('The average rental duration for movies rated NC-17 is {}.'.format(average[1]))
print('The average rental duration for movies rated PG is {}.'.format(average[2]))
print('The average rental duration for movies rated PG-13 is {}.'.format(average[3]))
print('The average rental duration for movies rated G is {}.'.format(average[4]))
```

### 3. What is the mean movie length? Calculate this by defining a function.

```python
# Construct the SQL statement
sql = '''
SELECT length
FROM film
'''

results = engine.execute(sql)
engine.dispose()
rows = results.fetchall()

# Extract the counts and print results
lengths = [row['length'] for row in rows]

# Define mean function
def mean(numbers):
    total_sum = 0 # Initialize the sum to zero
    for n in numbers:
        total_sum += n # Add up the numbers in the array
    count = len(numbers) # Find the length of the array
    avg = total_sum / count # Calculate the mean
    return avg # Return the result

# Return mean movie length
print('The mean movie length is {}.'.format(mean(lengths)))
```

### 4. What is the median movie length? Calculate this by defining a function.

```python
# Construct the SQL statement
sql = '''
SELECT length
FROM film
'''

results = engine.execute(sql)
engine.dispose()
rows = results.fetchall()

# Extract the counts and print results
lengths = [row['length'] for row in rows]

# Define median function
def median(numbers):
    numbers.sort() # Sort the list of numbers
    count = len(numbers) # Get the length of the array
    isEven = count % 2 == 0 # Check if this list is of even length

    if (isEven):
        # Find the two numbers in the middle of the array
        mid = math.floor(count / 2)
        a = numbers[mid - 1]
        b = numbers[mid]
        # Find the average of these two numbers
        ans = (a + b) / 2
    else:
        ans = numbers[math.floor(count / 2)]
    return ans

# Return mean movie length
print('The median movie length is {}.'.format(median(lengths)))
```

### 5. Calculate the standard deviation of the movie lengths. Calculate this my defining a function.

```python
# Construct the SQL statement
sql = '''
SELECT length
FROM film
'''

results = engine.execute(sql)
engine.dispose()
rows = results.fetchall()

# Extract the counts and print results
lengths = [row['length'] for row in rows]

# Define stdev function
def standard_deviation(numbers):
    X_bar = mean(numbers)
    N = len(numbers)
    total_sum = 0
    for X in numbers:
        diff = X - X_bar
        squared = math.pow(diff, 2)
        total_sum += squared
    sigma = math.sqrt(total_sum / (N - 1))
    return sigma

# Return mean movie length
print('The standard deviation of the movie length is {}.'.format(standard_deviation(lengths)))
```
